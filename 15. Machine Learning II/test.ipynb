```python
vid = cv2.VideoCapture(0)
if not vid.isOpened():
    print("No Camera")
    exit()

while True:
    res, frame = vid.read()
    if not res:
        break

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    cv2.imshow("Camera", rgb_frame)
    if cv2.waitKey(1) == ord('q'):
        break

vid.release()
cv2.destroyAllWindows()
```
