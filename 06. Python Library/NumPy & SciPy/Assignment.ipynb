### Goal

Use NumPy **indexing, slicing, and loops** to “draw” on a grayscale image.

### Tasks

1. Create a **300 × 300** NumPy array of zeros (dtype `float`), representing a **grayscale image**.
2. Using **slicing**, draw a **white square** in the center of the image:

   * Side length: e.g. 100 pixels.
3. Using **loops**, draw a **white diagonal line** from the top-left to the bottom-right.
4. Display the final image with `matplotlib.pyplot.imshow(..., cmap="gray")`.


````python
WIDTH, HEIGHT = 300, 300
SIDE = 100
START_X = (WIDTH - SIDE) // 2  # Same Y
END_X = START_X + SIDE   # Same Y


# Create Img 
img = np.zeros((WIDTH, HEIGHT), dtype = np.float32)

# Draw Squre 

img[START_X:END_X, START_X: END_X ] = 1

# Draw Line in Diagnoal
for idx in range(WIDTH):
    img[idx, idx] = 1


plt.imshow(img, cmap = 'gray')


````

    Assignment 2 — Solve this system of linear equations using numpy and scipy
</p>



```
x + 2y - 3z + 2w = 30
2x - 5y + 4z + 9w = 4
-5x + 40y - z - 20w = -6
5x - 4y - z + 60w = 5
```



````python
# Cofficient Matrix
A = np.array([
    [1, 2, -3, 2], 
    [2, -5, 4, 9],
    [-5, 40, -1, -20], 
    [5, -4, -1, 60]
])

# Result Matric
b = np.array([30, 4, -6, 5])
````

````python
# Solution By Numpy
X = np.linalg.solve(A, b)
print(tabulate.tabulate([X], headers = ('x', 'y', 'z', 'w'), tablefmt='fancy_grid'))
````

````python
# Solution By Numpy Inv
A_inv = np.linalg.inv(A)

# Get Solution by formula -> X = A_INT @ b
X = A_inv @ b
print(tabulate.tabulate([X], headers = ('x', 'y', 'z', 'w'), tablefmt='fancy_grid'))
````

````python
# Solution By Scipy 
X = sp.linalg.solve(A, b)
print(tabulate.tabulate([X], headers = ('x', 'y', 'z', 'w'), tablefmt='fancy_grid'))
````

## ✔ Deliverables

* Implement all four functions.
* Test them on small arrays (e.g., 5-10 elements).
* Confirm loop and vectorized versions produce identical results.



````python
def l1_loss_loops(y, y_pred):
    """
    Compute L1 loss using pure Python for-loops.
    y: NumPy array of true values
    y_pred: NumPy array of predicted values
    Returns: scalar L1 loss
    """
    loss = 0
    for idx in range(y.size):
        loss += np.absolute(y[idx] - y_pred[idx])
    return loss
    pass


def l1_loss_vectorized(y, y_pred):
    """
    Compute L1 loss using NumPy vectorized operations (NO LOOPS).
    Returns: scalar L1 loss
    """
    return np.sum(np.absolute(y - y_pred))
````

````python
def l2_loss_loops(y, y_pred):
    """
    Compute L2 loss using pure Python for-loops.
    y: NumPy array of true values
    y_pred: NumPy array of predicted values
    Returns: scalar L2 loss
    """
    loss = 0
    for idx in range(y.size):
        loss += (y[idx] - y_pred[idx]) ** 2
    return loss


def l2_loss_vectorized(y, y_pred):
    """
    Compute L2 loss using NumPy vectorized operations (NO LOOPS).
    Returns: scalar L2 loss
    """
    return np.sum(np.power(y - y_pred, 2))
    
````

````python
y = np.array([1, 2, 3], dtype=float)
y_pred = np.array([2, 1, 5], dtype=float)

# |1-2| + |2-1| + |3-5| = 1 + 1 + 2 = 4
print('#'*10, 'L1 Loss', '#'*10)
print("loops:", l1_loss_loops(y, y_pred))           
print("vectorized:", l1_loss_vectorized(y, y_pred)) 
print()

# (1-2)^2 + (2-1)^2 + (3-4)^2 = 1 + 1 + 1 = 3
print('#'*10, 'L2 Loss', '#'*10)
print("loops:", l2_loss_loops(y, y_pred))           
print("vectorized:", l2_loss_vectorized(y, y_pred)) 
print()
````
