```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("tab10")
sns.set_style('whitegrid')
```

<div style="text-align: center; background: #1ED760; font-family: 'Trebuchet MS', Arial, sans-serif; color: white; padding: 15px; font-size: 26px; font-weight: bold; line-height: 1; border-radius: 50% 0 50% 0 / 40px; margin-bottom: 20px; box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1);">Read Data</div>


```python
data = pd.read_csv("database.csv")
data.head()
```

<div style="text-align: center; background: #1ED760; font-family: 'Trebuchet MS', Arial, sans-serif; color: white; padding: 15px; font-size: 26px; font-weight: bold; line-height: 1; border-radius: 50% 0 50% 0 / 40px; margin-bottom: 20px; box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1);">Convert Date To DateTime</div>


```python
data['Date'] = data['Date'].apply(lambda x: pd.to_datetime(x.split('T')[0], format = "mixed"))
```

<div style="text-align: center; background: #1ED760; font-family: 'Trebuchet MS', Arial, sans-serif; color: white; padding: 15px; font-size: 26px; font-weight: bold; line-height: 1; border-radius: 50% 0 50% 0 / 40px; margin-bottom: 20px; box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1);">Month Vs Magnitude</div>


```python
data['Month'] = data['Date'].dt.month
month_grouped = data.groupby('Month')
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>1 | Using Seaborn</b></span>

```python
sns.lineplot(data, x = 'Month', y = 'Magnitude');
orig_ylim_min, orig_ylim_max = plt.ylim()
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>2 | Using Matplotlib</b></span>

```python
x = data['Month'].unique()
y = month_grouped['Magnitude'].mean().values

plt.plot(x, y, linewidth=2)
plt.fill_between(x, y - 0.02, y + 0.02, alpha=0.2)
plt.ylim(orig_ylim_min, orig_ylim_max)        
plt.xlabel("Month")
plt.ylabel("Magnitude")
plt.title("Month vs Magnitude")
plt.grid(alpha=0.3)

plt.show()
```

<div style="text-align: center; background: #1ED760; font-family: 'Trebuchet MS', Arial, sans-serif; color: white; padding: 15px; font-size: 26px; font-weight: bold; line-height: 1; border-radius: 50% 0 50% 0 / 40px; margin-bottom: 20px; box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1);">Year Vs Magnitude</div>


```python
data['Year'] = data['Date'].dt.year
year_grouped = data.groupby('Year')
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>1 | Using Seaborn</b></span>

```python
# using seaborn
sns.lineplot(data, x = 'Year', y = 'Magnitude')
orig_ylim_min, orig_ylim_max = plt.ylim()
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>2 | Using Matplotlib</b></span>

```python
x = data['Year'].unique()
y = year_grouped['Magnitude'].mean().values

plt.plot(x, y, linewidth=2)
plt.fill_between(x, y - 0.04, y + 0.04, alpha=0.2)
plt.ylim(orig_ylim_min, orig_ylim_max)        
plt.xlabel("Month")
plt.ylabel("Magnitude")
plt.title("Year vs Magnitude")
plt.grid(alpha=0.3)

plt.show()
```
