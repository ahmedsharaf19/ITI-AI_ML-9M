> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>1 | Read Data</b></span>

```python
tele_data = pd.read_csv("needed_files\WA_Fn-UseC_-Telco-Customer-Churn.csv")
tele_data.head()
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>2 | Set CustomerID as index</b></span>

```python
tele_data = tele_data.set_index('customerID')
tele_data.sample(1)
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>3 | Convert TotalCharges to numeric</b></span>

```python
tele_data['TotalCharges'] = pd.to_numeric(tele_data['TotalCharges'], errors = 'coerce')
tele_data['TotalCharges'].dtype
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>4 | Fill Na with Mean</b></span>

```python
avg_of_total_charges = tele_data['TotalCharges'].mean()
tele_data['TotalCharges'] = tele_data['TotalCharges'].fillna(avg_of_total_charges)
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>5 | Getting Cat Columns & Num Columns</b></span>

```python
# Get All Categorical Features
cat_columns = [c for c in tele_data.columns if tele_data[c].dtype == 'O' ]

# Get All Numerical Featues
nums_columns = [c for c in tele_data.columns if tele_data[c].dtype != 'O' ]

```

```python
cat_columns.remove('Contract')
cat_columns.remove("Churn")
cols_to_ohe = cat_columns

cols_to_label_encoding = ['Contract', 'Churn']

print("Columns To One Hot Encoding \n", cols_to_ohe)
print("Columns To Label Encoding \n", cols_to_label_encoding)
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>6 | Creating One Hot Encoding</b></span>

```python
one_hot_encoded_feat = pd.get_dummies(tele_data[cat_columns]).astype('int')
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>7 | Create Label Encoding For Contract</b></span>

```python
# Contract Column
mapping = {v: i+1 for i, v in enumerate(tele_data['Contract'].unique())}
print("Mapping Data \n", mapping)

tele_data['Contract'] = pd.to_numeric(tele_data['Contract'].map(mapping))
tele_data['Contract'].unique()
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>8 | Create Label Encoding For Churn</b></span>

```python
# Churn Columns
mapping = {v: i for i, v in enumerate(tele_data['Churn'].unique())}
print("Mapping Data \n", mapping)

tele_data['Churn'] = pd.to_numeric(tele_data['Churn'].map(mapping))
tele_data['Churn'].unique()
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>9 | Categorical Dataframe</b></span>

```python
cat_Df = pd.concat([one_hot_encoded_feat, tele_data[['Contract', "Churn"]]], axis = 1)
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>10 | Standardized Numerical Dataframe</b></span>

```python
num_Df = tele_data[nums_columns]
standardized_df = (num_Df - num_Df.mean(axis = 0)) / num_Df.std(axis = 0)
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>11 | Cleared DataFrame</b></span>

```python
cleared_data = pd.concat([num_Df, cat_Df], axis = 1)
cleared_data
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>12 | Splitting Data into Label/Data</b></span>

```python
label = cleared_data.reset_index()[['customerID', "Churn"]].set_index('customerID')
data = cleared_data.iloc[:, :-1]
```

> <span style='font-size:15px; font-family:Verdana;color: #FF00CC;'><b>13 | Saving</b></span>

```python
label.to_csv("label.csv")
data.to_csv("data.csv")
```
