Fill an array of 5 elements from the user, Sort it in descending 
and ascending orders then display the output.

```python
array = []
i = 0
while i < 5:
    num = input(f"Enter Number {i+1}: ")
    if not num.isdigit():
        print('Invalid Number Please Enter Integer Number Only')
        continue

    array.append(int(num))
    i += 1

array.sort()
print(array)
print(array[::-1])
```

* Write a program that generate a multiplication table from 1 to the 
number passed.


```python
number = input("Please Enter Number : ")
table = []
if number.isdigit():
    curr_num = 1
    while curr_num <= int(number):
        table.append(list())
        for i in range(1, curr_num+1):
            table[curr_num-1].append(curr_num*i) 
        curr_num += 1
    print(table)
else :
    print("Number Must Be Integer")
```
