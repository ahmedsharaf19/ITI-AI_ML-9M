* Write a program that counts up the number of vowels [a, e, i, o, 
u] contained in the string.

```python
# vowels char
vowels = "aeiou"
content = input("Please Enter Your String : ")
print("You Entered String : ", content)
counts = 0
for char in content:
    if char.lower() in vowels:
        counts += 1

print("Number Of Vowels is = ", counts)
```

* Write a program that prints the number of times the string 'iti' 
occurs in anystring.

```python
iti = "iti"
content = input("Please Enter Your String : ")
print("You Entered String : ", content)
counts = 0
for word in content.split():
    if word.lower() == iti:
        counts += 1
print("Number of iti occurence is = ", counts)
```

* Write a program that prints the locations of "i" character in any 
string you added.

```python
content = input("Please Enter Your String : ")
print("You Entered String : ", content)
idx = 0
line = "i in idx : "
for char in content:
    if char == 'i':
        line += str(idx) +' '
    idx += 1
    
print(line)
```

* Write a program that generate a multiplication table from 1 to the 
number passed.


```python
# 1   -> 1
# 2   -> 2, 4
# 3   -> 3, 6, 9
# 4   -> 4, 8, 12, 16
```

```python
number = input("Please Enter Number : ")
line = ""
if number.isdigit():
    curr_num = 1
    while curr_num <= int(number):
        for i in range(1, curr_num+1):
            line += str(curr_num * i) +' '
        line += '\n'
        curr_num += 1
    print(line)
else :
    print("Number Must Be Integer")
```

* Write a program that build a Mario pyramid like below:

```python
# 4 
#___*   i = 0, j = 3        j >= 4 - 0 - 1 = 3
#__**   i = 1, j >= 2       j >= 4 - 1 - 1 = 2
#_***   i = 2, j >= 1       j >= 4 - 2 - 1 = 1
#****   i = 3, j >= 0       j >= 4 - 3 - 1 = 0

# so conition is j >= number - i - 1
```

```python
number = input("Please Enter Number : ")
line = ""
i = 0
if number.isdigit():
    number = int(number)
    while i < number:
        j = 0
        while j < number:
            if j >= number - i - 1:
                line += '*'
            else:
                line += ' '
            j += 1
        i += 1
        line += '\n'
else:
    print("Number Must Ingeger")

print(line)
```
