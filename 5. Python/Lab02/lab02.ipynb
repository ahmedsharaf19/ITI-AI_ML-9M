* Write a function that accepts two arguments (length, start) to 
generate an array of a specific length filled with integer numbers 
increased by one from start

```python
def create_array(length, start):
    if length < 0:
        return 
    else :
        return list(range(start, start+length))

length = input("Enter Length : ")
start = input("Enter Start : ")

if length.isdigit() and start.isdigit():
    array = create_array(int(length), int(start))
    if array:
        print('Array is : ', array)
    else : 
        print('Length Can not be negative')

else :
    print("Length & Start Must Be Integer Value")
```

* write a function that takes a number as an argument and if the 
numberdivisible by 3 return "Fizz" and if it is divisible by 5 return 
"buzz" and if is is divisible by both return "FizzBuzz"

```python
def divisability(num):
    result = ''
    if num % 3 == 0:
        result += 'Fizz'
    
    if num % 5 == 0:
        result += 'buzz'
    
    return result

print(divisability(15))
print(divisability(3))
print(divisability(9))

```

- Write a function which has an input of a string from user then it 
will return the same string reversed.

```python
def reversedSlicing(string):
    return string[::-1]

def reverseIterative(string):
    reversed = ''
    for i in range(-1,-len(string)-1, -1):
        reversed += string[i]
    return reversed
    
print(reversedSlicing("Ali"))
print(reverseIterative('ali'))
```

* Ask the user for his name then confirm that he has entered his 
name(not an empty string/integers). then proceed to ask him for 
his email and print all this data(Bonus) check if it is a valid email 
or not

```python
import re
email_pattern = re.compile(r'[\w.]+@[a-zA-Z]+.[a-zA-Z]+')

while True:
    name = input("Enter your name: ").strip()
    if name.replace(' ', '').isalpha():
        break
    print("Invalid name. Please enter a non-empty name containing letters only.")
    
email = input("Enter your email: ").strip()
is_valid = bool(email_pattern.match(email))

print()
print("Collected data:")
print("Name :", name)
print("Email:", email)
print("Email valid:" , is_valid)
```

* Write a function that takes a string and prints the 
longest alphabetical ordered substring occurred For example, if 
the string is 'abdulrahman' then the output is: Longest substring in 
alphabetical order is: abdu


```python
def long_sub_string(string):
    longest = current = string[0]

    for i in range(1, len(string)):
        if string[i] >= string[i - 1]:
            current += string[i]
        else:
            current = string[i]

        if len(current) > len(longest):
            longest = current
    return longest

print('Longest substring in alphabetical order is:', long_sub_string("abdulrahman"))
```

* Write a program which repeatedly reads numbers until the user 
enters “done”.
 ○ Once “done” is entered, print out the total, count, and 
average of the numbers.
 ○ If the user enters anything other than a number, detect their 
mistake, print an error message and skip to the next number

```python
inp = ''
numbers = []
total = 0
count = 0
while inp.lower() != 'done':
    inp = input("Enter Number : ")
    if not inp.isdigit():
        print('Please Enter Only Integer Numbers!')
        continue
    count += 1
    total += int(inp)

print("Counting Number is : ", count)
print("Total Of Number is : ", total)
print("Average Number is : ", 0 if count == 0 else total / count)
```

* Word guessing game (hangman)
    -  A list of words will be hardcoded in your program, out of which the interpreter will
    -  choose 1 random word.
    -  The user first must input their names
    -  Ask the user to guess any alphabet. If the random word  contains that alphabet, it
    -  will be shown as the output(with correct placement)
    -  Else the program will ask you to guess another alphabet.
    -  Give 7 turns maximum to guess the complete word

```python
import random

words = ['python', 'apple', 'match', 'sharaf', 'ahmed', 'elsherbiny', 'iti']
word = random.choice(words)
display = ["_"] * len(word)
turns = 7

name = input("Enter your name: ").strip()
print(f"Welcome, {name}!")
print("Guess the word:", " ".join(display))

while turns > 0 and "_" in display:
    guess = input("Guess a letter: ").lower()
    if len(guess) != 1 or not guess.isalpha():
        print("Please enter one letter only")
        continue
    
    found = False
    for i, ch in enumerate(word):
        if ch == guess and display[i] == "_":
            display[i] = guess
            found = True
            break

    if found:
        print("Correct:", " ".join(display))
    else:
        turns -= 1
        print(f"Wrong, You have {turns} turns left")
        print(" ".join(display))

if "_" not in display:
    print(f"Congrats {name}! You guessed the word: {word}")
else:
    print(f"Game over, The word was: {word}")
```
