```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset("iris")
data.head()
```

```python
data.boxplot()
```

```python
Q3 = data['sepal_width'].quantile(0.75)
Q1 = data['sepal_width'].quantile(0.25)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
data = data[(data['sepal_width'] <= upper_bound) & (data['sepal_width'] >= lower_bound)]
```

```python
mean = data['sepal_width'].mean()
std = data['sepal_width'].std()
data = data[(data['sepal_width'] >= mean - 3 * std) & (data['sepal_width'] <= mean + 3 * std)]
```
